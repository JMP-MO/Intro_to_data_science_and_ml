### What is Machine Learning?

Machine learning is **predicting a number, with some other numbers** using an **algorithm that learns from data**. For instance:
> 1. If we trained a model on historic weather data and know the temperature now, could we predict it in 6 hours? (Regression)
> 2. If we represent words with numbers and trained a model on historic text and text on the internet, if given a sequence of words (numbers), can we predict the next words (numbers).
> 3. If we trained a model by giving it millions of labelled images, could we predict the label associated with the key object in the image. (Supervised learning) (Classification)
> 4. If given unlablled data such as medical records, could we train a model to recognise patterns that may indicate the patient has a high chance of developing a particular disease. (Unsupervised learning) (Clustering)


There are many more types of machine learning but these are the main categories you probably want to start with:
* Supervised and Unsupervised learning
* Regression and Classification

Lets run a little experiment to show the speed difference using numpy over a python list.
>Do you know WHY numpy can execute a repetitive function such as this so much quicker?

```python
import time

size = 10_000_000
py_list = [x for x in range(size)]  # Create a python list - using 'list comprehension' loop for extra speed 
np_array = np.array(py_list)        # Create a numpy array from the py_list.

# --- Python list ---
start = time.time()
total_list = sum([x**2 for x in py_list])
end = time.time()
print(f"Python list time: {end - start:.5f} seconds")

# --- NumPy array ---
start = time.time()
total_numpy = np.sum(np_array**2)
end = time.time()
print(f"NumPy array time: {end - start:.5f} seconds")
```
